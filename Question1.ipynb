```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde

# Generate some random data (for demonstration)
np.random.seed(0)
data = np.random.normal(0, 1, size=1000)

# Perform Kernel Density Estimation
kde = gaussian_kde(data, bw_method=0.2)  # bw_method controls the bandwidth (smoothing)

# Create a range of values over which to evaluate the KDE
x = np.linspace(-5, 5, 1000)

# Evaluate the KDE on the range of x values
kde_values = kde(x)

# Plot the results
plt.figure(figsize=(8, 6))
plt.plot(x, kde_values, label='KDE', color='blue', lw=2)
plt.hist(data, bins=30, density=True, alpha=0.5, color='gray', label='Histogram')
plt.title('Kernel Density Estimation')
plt.xlabel('Data values')
plt.ylabel('Density')
plt.legend()
plt.show()
```
